```python
import yfinance as yf
import pandas as pd
import os

# Asset selection
data = pd.read_html("https://en.wikipedia.org/wiki/S%26P_100")
tickers = data[2]["Symbol"].tolist()
tickers[tickers.index("BRK.B")] = "BRK-B"
```

```python
# Import 10 years of historic data to store in data/raw using yfinance to fetch OHLCV (Open, High, Low, Close, Volume) data
import_data = yf.download(tickers, start = "2015-01-01", end = "2024-12-31", auto_adjust = False)
import_data.to_csv(os.path.join("..", "data", "raw", "import_data.csv"))

for ticker in tickers:
    dataframe = import_data.xs(ticker, level = 1, axis = 1)
    dataframe.to_csv(os.path.join("..", "data", "raw", f"{ticker}.csv"))
```

```python
# Pre-processing data, handle missing values/non-trading days

datas = os.path.join("..", "data", "raw")
cleaned_data_path = os.path.join("..", "data", "processed")

for file in os.listdir(datas):
    file_path = os.path.join(datas, file)
    # check for missing values
    single_column = pd.read_csv(file_path)
    single_column.fillna(method = 'ffill', inplace = True)  
    single_column.fillna(method = 'bfill', inplace = True)
    single_column.to_csv(os.path.join(cleaned_data_path, file), index = False)
```

```python
# Feature engineering (moving averages, volatility, RSI, etc.)
# Save cleaned data to /data/processed/

for file in os.listdir(cleaned_data_path):
    file_path = os.path.join(cleaned_data_path, file)
    if os.path.isdir(file_path):
        continue  
    single_column = pd.read_csv(file_path)
    single_column['Adj Close'] = pd.to_numeric(single_column['Adj Close'], errors = 'coerce')
    single_column['10_MA'] = single_column['Adj Close'].rolling(window = 10).mean()
    single_column['Volatility'] = single_column['Adj Close'].pct_change().rolling(window = 10).std()
    single_column.to_csv(file_path, index=False) 
```
